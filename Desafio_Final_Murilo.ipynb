```python
##### PREPARANDO DADOS DA COLUNA TEMPO PARA IMPRIMIR GRÁFICO #####

# Criando dicionário a partir dos valores de tempo presentes nos dados
dicionario_tempo = dict(df.tempo.value_counts())

print ("Olá Joana, os valores da coluna 'tempo' presentes na tabela são:\n")

# ordenando os numeros em ordem crescente e imprimindo-os

for item in sorted(dicionario_tempo): 
    print("- %d horas, que aparece %d vezes" % (item, dicionario_tempo[item]))
    
    
print("\nNo gráfico abaixo fica mais fácil de visualizar:")
    
    

    
######### GRÁFICO DE BARRAS #########

# Definindo função pra posicionar valores acima das barras no gráfico
# esta função foi modificada a partir da original, disponível em: https://stackoverflow.com/a/56780852

def show_values_on_bars(axs, h_v="v", space=0.4, size = 12.0, color='black'):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + float(space)
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center", size=size, color=color) 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left", size=size, color=color)

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
        
# convertendo o dicionario com os dados de tempo para um dataframe do pandas
data_tempo = pd.DataFrame.from_dict(dicionario_tempo.items())

# criando e imprimindo o grafico de barras no seaborn

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=0, y=1, data=data_tempo , palette = "Purples", order = [24, 48, 72])
ax.set_title("Janelas de tempo".upper(), pad=4.0, fontsize=16, fontweight = 'bold', color = 'dimgrey', y = 1.05)
ax.set_xlabel('Tempo (horas)', labelpad = 5.0, fontsize=14,  color = 'dimgrey')
ax.set_ylabel('Quantidade', labelpad = 10.0, fontsize=14, color = 'dimgrey')
ax.set_ylim(ymin=7000)
show_values_on_bars(ax, size = 15, space = 30, color = 'dimgrey')
plt.xticks(rotation = 0, fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

### Modelos e Resultados

— Agora vamos rodar 4 modelos diferentes de aprendizado de máquina, todos bastante simples, pra ver se ele consegue adivinhar os experimentos que vão despertar ou não uma propriedade anti-inflamatória. Cada modelo vai "ensinar" o computador de um jeito diferente. É como se cada um fosse um óculos diferente, então pra cada óculos que ele usar, vai "enxergar" essa tabela de um jeito um pouquinho diferente. E isso fará ele nós dar respostas diferentes.

```python
# separando conjunto de dados entre teste e treino

from sklearn.model_selection import train_test_split

x = df_antiinflamatorio.drop(['id', 'droga', 'anti-inflammatory'], axis=1)
x = pd.get_dummies(x, columns=['tratamento', 'tempo', 'dose'])
y = df_antiinflamatorio['anti-inflammatory']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3, stratify=y, random_state = 42)
```

```python
## criando dicionario pra guardar os resultados dos testes
resultados = {}
```

```python
##### Rodando modelo 1: LogisticRegression

from sklearn.linear_model import LogisticRegression

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
resultado_logistic_reg = modelo_rlogistica.score(x_teste, y_teste)

resultados["LogisticRegression"] = resultado_logistic_reg
resultado_logistic_reg
```

```python
##### Rodando modelo 2: RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
resultado_randomforest = modelo_randomforest.score(x_teste, y_teste)

resultados["RandomForest"] = resultado_randomforest
resultado_randomforest
```

```python
##### Rodando modelo 3: LinearSVC

from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification

#X, y = make_classification(random_state=0)

modelo_linearSVC = make_pipeline(StandardScaler(),
                    LinearSVC(random_state=0, max_iter=5000))
modelo_linearSVC.fit(x_treino, y_treino)
resultado_linearSVC = modelo_linearSVC.score(x_teste, y_teste)

resultados["LinearSVC"] = resultado_linearSVC
resultado_linearSVC
```

```python
##### Rodando modelo 4: Dummy
## strategy = most_frequent

from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier(strategy="most_frequent") # generates predictions by respecting the training set’s class distribution.
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
resultado_dummy = accuracy_score(y_teste, previsao_dummy)

resultados["Dummy"] = resultado_dummy
resultado_dummy
```

```python
# Preparando dados a serem usados nos gráficos

# convertendo o dicionario com os dados de tempo para um dataframe do pandas
resultados_ML = pd.DataFrame.from_dict(resultados.items())
resultados_ML = resultados_ML.rename({0: 'modelo', 1: 'taxa_acerto'}, axis =1)
```

```python
######### GRÁFICOS DE BARRAS: ACURÁCIA DOS MODELOS #########

# Definindo função pra posicionar valores acima das barras no gráfico
# esta função foi modificada a partir da original, disponível em: https://stackoverflow.com/a/56780852

def show_values_on_bars_float(axs, h_v="v", space=0.4, size = 12.0, color=color):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + float(space)
                value = (p.get_height())
                value = "{:.2%}".format(value)
                ax.text(_x, _y, value, ha="center", size=size, color=color) 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left", size=size, color=color)

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
        



# criando e imprimindo os graficos de barras no seaborn

#### GRAFICO 1 - ESCALA 0 ~1 ####

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='modelo', y='taxa_acerto', data=resultados_ML , palette = "Purples")#, order = [24, 48, 72])
ax.set_title("Acurácia dos modelos".upper(), pad=4.0, fontsize=16, fontweight = 'bold', color = 'dimgrey', y = 1.05)
ax.set_xlabel('Nome do Modelo', labelpad = 10.0, fontsize=14,  color = 'dimgrey')
ax.set_ylabel('Taxa de Acertos (de 0 a 1)', labelpad = 10.0, fontsize=14, color = 'dimgrey')
ax.set_ylim(ymin=0, ymax= 1)
#show_values_on_bars_float(ax, size = 15, space = 30, color = 'dimgrey')
plt.xticks(rotation = 0, fontsize=12)
plt.yticks(fontsize=14)
plt.show()
```

```python
#### GRAFICO 2 - ESCALA 0.99 ~ 1 ####

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='modelo', y='taxa_acerto', data=resultados_ML , palette = "Purples")#, order = [24, 48, 72])
ax.set_title("Acurácia dos modelos".upper() + "\n(Escala Ampliada)", pad=4.0, fontsize=16, fontweight = 'bold', color = 'dimgrey', y = 1.05)
ax.set_xlabel('Nome do Modelo', labelpad = 10.0, fontsize=14,  color = 'dimgrey')
ax.set_ylabel('Taxa de Acertos (de 0.99 a 1)', labelpad = 10.0, fontsize=14, color = 'dimgrey')
ax.set_ylim(ymin=0.99, ymax= 1)
show_values_on_bars_float(ax, size = 14,space = 0, color = 'dimgrey')
plt.xticks(rotation = 0, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
